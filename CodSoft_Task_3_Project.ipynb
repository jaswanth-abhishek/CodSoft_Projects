# Impact of TV, Radio, and Newspaper Advertising on Sales: Scatter and Regression Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x='TV',y='Sales',data=advertising_data,scatter_kws={'color':'blue'},line_kws={'color':'red'})
sns.lmplot(x='Radio',y='Sales',data=advertising_data,scatter_kws={'color':'green'},line_kws={'color':'black'})
sns.lmplot(x='Newspaper',y='Sales',data=advertising_data,scatter_kws={'color':'purple'},line_kws={'color':'orange'})
plt.show()
```

# Distribution of Advertising Spend Across Different Media

```python
fig,axs=plt.subplots(1,3,figsize=(15, 5))
sns.boxplot(y=advertising_data['TV'],color='orange',ax=axs[0])
axs[0].set_title('TV Advertising Spend')
axs[0].set_ylabel('')
sns.boxplot(y=advertising_data['Radio'],color='purple',ax=axs[1])
axs[1].set_title('Radio Advertising Spend')
axs[1].set_ylabel('')
sns.boxplot(y=advertising_data['Newspaper'],color='cyan',ax=axs[2])
axs[2].set_title('Newspaper Advertising Spend')
axs[2].set_ylabel('')
plt.tight_layout()
plt.suptitle('Distribution of Advertising Spend Across Different Media')
plt.show()
```

```python
fig,axs=plt.subplots(1,3,figsize=(15,5))
sns.histplot(advertising_data['TV'],color='orange',ax=axs[0],kde=True)
axs[0].set_title('Distribution of TV Advertising Spend')
sns.histplot(advertising_data['Radio'],color='purple',ax=axs[1],kde=True)
axs[1].set_title('Distribution of Radio Advertising Spend')
sns.histplot(advertising_data['Newspaper'],color='cyan',ax=axs[2],kde=True)
axs[2].set_title('Distribution of Newspaper Advertising Spend')
plt.tight_layout()
plt.suptitle('Distribution of Advertising Spend Across Different Media')
plt.show()
```

# Correlation Heatmap of Advertising Spend

```python
ad_corr_matrix=advertising_data[['Radio','Newspaper','TV']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(ad_corr_matrix,annot=True,cmap='coolwarm',center=0,fmt='.2f',linewidths=0.5,linecolor='black',cbar_kws={'shrink':0.8})
plt.title('Correlation Heatmap of Advertising Spend')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
x=advertising_data[['Radio','TV','Newspaper']]
y=advertising_data[['Sales']]
```

```python
import statsmodels.api as sm
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print(f'Coefficients:{model.coef_}')
print(f'Intercept:{model.intercept_}')
print(f'Mean Squared Error:{mean_squared_error(y_test, y_pred)}')
print(f'R-squared:{r2_score(y_test, y_pred)}')
```

```python
x2_test=sm.add_constant(x)
reg=sm.OLS(y,x2_test)
model=reg.fit()
print(model.summary())
```
